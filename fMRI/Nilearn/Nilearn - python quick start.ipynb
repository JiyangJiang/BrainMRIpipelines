```python
rsn = datasets.fetch_atlas_smith_2009()['rsn10']
print (rsn)
```

```python
from nilearn import image
print (image.load_img (rsn).shape)
```

```python
first_rsn = image.index_img (rsn, 0)
print (first_rsn.shape)
```

```python
plotting.plot_stat_map (first_rsn, threshold = 3)
```

```python
for idx in range (0,9,1):
    curr_rsn = image.index_img (rsn, idx)
    plotting.plot_stat_map (curr_rsn, threshold = 3)
```

```python
for img in image.iter_img (rsn):
    plotting.plot_stat_map (img, threshold = 3, display_mode = "z", cut_coords = 1, colorbar = False)
plotting.show()
```
