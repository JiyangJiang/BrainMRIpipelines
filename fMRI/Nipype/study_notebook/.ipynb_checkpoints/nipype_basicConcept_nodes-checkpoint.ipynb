```python
# a node is an object to execute a certain function.
# This function can be an interface, a user-defined function, or external script
# Each node consists of a name, an interface category, and at least one input and one output fields
#
# In general, a node consists of the following elements
#
#     nodename = Nodetype (interface_function (), name = 'labelname')
#
# nodename : variable name of the node in python environment
# Nodetype : Node, MapNode, or JoinNode
# interface_function : can be user specific or coming from Interface
# labelname : label name of node in workflow environment (defines the name of working dir)
```

```python
# Example 1
from nipype import Node, Function

def add_two (x_input):
    return x_input + 2

addtwo = Node (Function (input_names = ['x_input'],
                         output_names = ['val_output'],
                         function = add_two),
               name = 'add_node')

addtwo.inputs.x_input = 4

temp_res = addtwo.run()
```

```python
temp_res.outputs
```

```python
addtwo.result.outputs
```

```python
from nipype.interfaces.fsl import BET
from nipype import Node

bet = Node (BET (frac = 0.3), name = 'bet_node')
bet.inputs.in_file = '/home/jiyang/Work/sub-2019A/anat/sub-2019A_T1w.nii.gz'
bet.inputs.out_file = '/home/jiyang/Work/sub-2019A/derivatives/sub-2019A_T1w_bet.nii.gz'

bet_res = bet.run()
```

```python
from nilearn.plotting import plot_anat
%matplotlib inline
import matplotlib.pyplot as plt
plot_anat (bet.inputs.in_file, title = 'BET input', cut_coords = (10,10,10),
           display_mode = 'ortho', dim = -1, draw_cross = False, annotate = False)
plot_anat (bet_res.outputs.out_file,  title = 'BET output', cut_coords = (10,10,10),
           display_mode = 'ortho', dim = -1, draw_cross = False, annotate = False)
```

```python
bet_res.outputs
```

```python
# Example 2

from nipype import Node
from nipype.interfaces.fsl import IsotropicSmooth

smooth = Node (IsotropicSmooth (), name = 'smoothnode')

smooth.inputs.in_file = '/home/jiyang/Work/sub-2019A/anat/sub-2019A_T1w.nii.gz'
smooth.inputs.out_file = '/home/jiyang/Work/sub-2019A/derivatives/sub-2019A_T1w_smooth.nii.gz'
smooth.inputs.fwhm = 4

smooth_res = smooth.run()
```

```python
from nilearn.plotting import plot_anat
%pylab inline
plot_anat (smooth.inputs.in_file, title = 'smooth input', cut_coords = (10,10,10),
           display_mode = 'ortho', dim = -1, draw_cross = False, annotate = False)
plot_anat (smooth_res.outputs.out_file, title = 'smooth output', cut_coords = (10,10,10),
           display_mode = 'ortho', dim = -1, draw_cross = False, annotate = False)
```
