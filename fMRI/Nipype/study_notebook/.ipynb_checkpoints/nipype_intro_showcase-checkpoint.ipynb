```python
# import the main Nipype tools - Node and Workflow
from nipype import Node, Workflow
```

```python
# import interfaces that we want to use for preprocessing
from nipype.interfaces.fsl import SliceTimer, MCFLIRT, Smooth
```

```python
# put three interfaces into nodes and define specific input paramters

# Initiate a node to correct for slice-wise acquisition (slice timing)
slicetimer = Node (SliceTimer (index_dir = False,
                               interleaved = True,
                               time_repetition = 2.5),
                   name = "slicetimer")

# Initiate a node to correct for motion
mcflirt = Node (MCFLIRT (mean_vol = True,
                         save_plots = True),
                name = "mcflirt")

# Initiate a node to smooth fMRI
smooth = Node (Smooth (fwhm = 4), name = "smooth")
```

```python
# After creating the nodes, now create preprocessing workflow

preproc01 = Workflow (name = 'preproc01', base_dir = ".")

# connect all nodes
preproc01.connect ([(slicetimer, mcflirt, [('slice_time_corrected_file', 'in_file')]),
                    (mcflirt,    smooth,  [('out_file',                  'in_file')])])
```

```python
# Visualise workflow
preproc01.write_graph (graph2use = 'orig')
```

```python
# Visualise workflow (contd)
from IPython.display import Image
Image (filename = "preproc01/graph_detailed.png")
```

```python
# ============================== #
# Apply the workflow to one fMRI #
# ============================== #

slicetimer.inputs.in_file = '/Users/z3402744/Work/test/sub-1805A/func/sub-1805A_task-rest_bold.nii.gz'

# use 5 processors to parallelise the processing, but since only one subject, it is not
# possible to parallelise. Therefore, it will run sequentially.
%time preproc01.run ('MultiProc', plugin_args = {'n_procs' : 5})
```

```python
!tree preproc01 -I '*js|*json|*pklz|_report|*.dot|*html'
```

```python
# Now change fwhm value of Gaussian smoothing kernel to 2, and rerun workflow
smooth.inputs.fwhm = 2;

%time preproc01.run ('MultiProc', plugin_args = {'n_procs' : 5})

# Nipype should skip slicetimer and mcflirt, because they didn't change from
# last execution.
```

```python
# Running workflow in parallel
# =============================

# copy/clone preproc01
preproc02 = preproc01.clone ('preproc02')
preproc03 = preproc01.clone ('preproc03')
preproc04 = preproc01.clone ('preproc04')
preproc05 = preproc01.clone ('preproc05')
```

```python
# we now have five preprocessing workflows. Put them all in another workflow to parallelise
metaflow = Workflow (name = 'metaflow', base_dir = '.')

# add the five preproc workflows into metaflow
metaflow.add_nodes ([preproc01,
                     preproc02,
                     preproc03,
                     preproc04,
                     preproc05])
# now we have metaflow which contains five other workflows (preproc0?), each of which further contains
# three nodes (slicetimer, mcflirt, smooth)
```

```python
# visualise metaflow
metaflow.write_graph (graph2use = 'flat')
from IPython.display import Image
Image (filename = "metaflow/graph_detailed.png")
```

```python
# parallel processing
# The following command process the same image (/Users/z3402744/Work/test/sub-1805A/func/sub-1805A_task-rest_bold.nii.gz)
# five times using parallel processing.
%time metaflow.run ('MultiProc', plugin_args = {'n_procs' : 5})
```
